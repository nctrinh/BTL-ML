```python
import matplotlib.pyplot as plt
```

## 1. Trên dữ liệu để huấn luyện và test

### 1.1. Chênh lêch Accuracy trên tập huấn luyện

```python
values = [94.86, 97.55, 94.29, 100]
legends = ["LG_without_optimization", "LG_fillNaN_by_algorithm", 
           "LG_fillNaN_by_algorithm(0)", "XGboost"]

plt.figure(figsize=(10, 6))
plt.bar(legends, values, color=['blue', 'orange', 'green', 'red'], width=0.5)

plt.title("Chênh lệch Accuracy trên tập huấn luyện", fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.xlabel("Models", fontsize=12)
plt.ylim(90, 101)

for i, v in enumerate(values):
    plt.text(i, v + 0.01, f"{v}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
